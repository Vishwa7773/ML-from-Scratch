```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

Naive bayes works on conditional probability  
P(Y=y | X=$x_{1},x_{2},x_{3}....x_{n}$)

$y^{'}$= arg max(P(Y=y | X))

Bayes theorum:  
P(Y/X)= (P(X/Y)*P(Y))/P(X)  
  
where,  
P(Y/X) = probablity of Y after observing features X.    
P(Y) = Probality of class Y after seeing any feature evidence.  
P(X/Y) = Likelihood. 
P(X) = Evidence.   



Assuming all X features as independent.  
P(X/Y)= P($X_{1}$/Y) * P($X_{2}$/Y) * P($X_{3}$/Y)...P($X_{n}$/Y)  
P(X/Y)=$\Pi_{i=1}^{n}$ $P(X_{i}/Y)$


```python
data=pd.read_csv(r'C:\Users\HP\OneDrive\Desktop\Machine Learning\0.Data\binomial_logistic_regression_dataset.csv')
data.head()
```

```python
data.rename(columns={'Pass(1)_Fail(0)':'Result'},inplace =True)
```

```python
x=data.drop(columns='Result')
y=data['Result']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.25)
```

```python
class NaiveBayes:
    def __init__(self):
        pass
    
    def fit(self,x_train,y_train):
        self.classes=np.unique(y_train)
        self.priors=[len(y_train[y_train==c])/len(y_train) for c in self.classes]
        self.means=[x_train[y_train==c].mean() for c in self.classes]
        self.std=[x_train[y_train==c].std() for c in self.classes]

    def likelihood(self,row,class_idx):
        likelihood=1
        for features in row.index:
            mean=self.means[class_idx][features]
            std=self.std[class_idx][features]
            std = max(std, 1e-9)
            likelihood *= ((1 / (np.sqrt(2*np.pi) * std))* np.exp(-((row[features] - mean)**2) / (2 * std**2)))
        
        return likelihood
    def predict(self,x):
        pred=[]

        for _,row in x.iterrows():
            posterior=[]
            for i in range(len(self.classes)):
                likelihood=self.likelihood(row,i)
                posterior.append(likelihood*self.priors[i])
            
            pred.append(np.argmax(posterior))

        return np.array(pred)       
```

```python
model=NaiveBayes()
model.fit(x_train,y_train)
predictions=model.predict(x_test)
accuracy=np.mean(predictions==y_test)
accuracy*100
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

model = GaussianNB()
model.fit(x_train, y_train)

pred = model.predict(x_test)

print(accuracy_score(y_test, pred)*100)
```
